```python
#Exercicio 2
compra= float(input("digite o valor de sua compra: "))

if compra <= 100:
    print("voce nao tem desconto!")
elif compra > 100:
    div= compra-(compra * 0.10)
    print(f"voce tem 10 de desconto! sua compra ficou com o total de {div} reais!")
elif compra <= 500:
    div= compra-(compra * 0.10)
    print(f"voce tem 10 de desconto! sua compra ficou com o total de {div} reais!")
else:
    divisao= compra(compra * 0.20)
    print(f"voce tem vinte de desconto! sua compra ficou com o total de {divisao} reais!")

```
