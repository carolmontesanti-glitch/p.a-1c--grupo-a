```python
#peça uma palavra e conte as letras
palavra = input("digite a palavra: ")
quantidade = 0
for i in range(len(palavra)):
  quantidade += 1
print(f"quantidade de letras {quantidade}")
```

```python
#peça uma palavra e conte as vogais
word = input("digite uma palavra")
vowel_count = sum(1 for char in word.lower() if char in "aeiu")
print(f"a palavra {word} possui {vowel_count} vogais")
```
